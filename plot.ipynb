```python
def parse_data(data_file):

    with open(data_file) as file:
        
        lines = file.readlines()

    a = ([line.strip().split() for line in lines])

# turn array strings into numbers
    
    b = np.array(a)
    u = b.astype(np.float)

    r = u[:,0] # radial distance
    c = u[:,1] # charge density

    return r, c
```

# Multiple plots in one graph

```python
def plot_data(files, xlabel, ylabel, colors, axeslimits = None, legends = None):
    
    fig = plt.figure()
    ax  = fig.add_subplot(111)
    
    count = 0
    
    if legends != None:
    
        for file, legend in zip(files, legends):

            r, c = parse_data(file)
        
            plt.plot(r, c, linewidth = 3, color = colors[count], label = legend)
        
            count += 1
            
        plt.legend(fontsize = 18, fancybox = True, \
               framealpha = 0.4, loc = 1)#, bbox_to_anchor=(0.8, -0.1))
    else:
        
        for file in files:

            r, c = parse_data(file)
        
            plt.plot(r, c, linewidth = 3, color = colors[count])
        
            count += 1
            
    plt.xlabel(xlabel, fontsize = 20)
    plt.ylabel(ylabel, fontsize = 20)
        
    if axeslimits != None:
        plt.axis(axeslimits)
    
    #ax.set_yscale('log', basey = np.e)
    
    plt.grid(True)
    
    ax.tick_params(which = 'major', direction = 'in', length = 8, top = True, right = True)
    ax.tick_params(which = 'minor', direction = 'in', length = 4, top = True, right = True)
    ax.tick_params(labelsize = 15, grid_alpha = 0.4, grid_linestyle = '--')
    ax.set_axisbelow(True)
    
    ax.xaxis.set_major_locator(MultipleLocator(0.5))
    ax.xaxis.set_minor_locator(AutoMinorLocator(5)) # for the minor ticks,
                                                    #use no labels; default NullFormatter
    
    ax.yaxis.set_major_locator(MultipleLocator(0.05))
    ax.yaxis.set_minor_locator(AutoMinorLocator(5)) 
    
    #majorFormatter = FormatStrFormatter('%d')
    #ax.xaxis.set_major_formatter(majorFormatter)
    
    plt.subplots_adjust(top = 1.1) #Make room for the ridiculously large title.
    #plt.suptitle(r"$\bf {30}$" \
    #r' ' r"$\bf { mM}$" , x = 0.5 , y = 1.3, fontsize = 24, color = 'k')
    
    #plt.savefig(common_path + 'images/mgden-33mM-soft-fit.png',\
                 #bbox_inches = 'tight', dpi = 200)
    #plt.show()
    
    return None
```
